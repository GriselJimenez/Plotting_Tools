```python
import os
import numpy as np
import logplot as lp
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
DF=pd.read_csv('WellData.dat')
DF.head()
```

```python
Well_ID=10
DF=DF[DF['WellID']==Well_ID]
```

```python
fig = plt.figure(figsize=(6,12), dpi=150, facecolor='w', edgecolor='k')

ax=fig.add_subplot(131)
lp.logplot(DF['Z'],DF['phie'], cat=False, ax=ax, 
           aspect='auto', varname='Porosity', lc='green')

ax=fig.add_subplot(132)
lp.logplot(DF['Z'],DF['kv'], cat=False, ax=ax, aspect='auto', 
           varname='Permeability (md)', lc='blue')

ax=fig.add_subplot(133)
List=[(204, 0, 0), (255, 208, 0), (255, 147, 0),(0, 204, 0),(153, 153, 153)]
namelist=['BRCC','SAND','SIHS','MIHS','MDST']
lp.logplot(DF['Z'],DF['Facies'], cat=True, ax=ax, aspect='auto', varname='Facies',
           colorlist=List,namelist=namelist,reversey=False,legend_fontsize=10)

plt.tight_layout(pad=1.08, h_pad=1.02, w_pad=1.2)
```
