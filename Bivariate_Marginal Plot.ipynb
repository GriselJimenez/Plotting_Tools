```python
import numpy as np
from matplotlib import pyplot as plt
import PythonTools as PT # My personal python libraray

%load_ext autoreload
%autoreload 2

%matplotlib inline
PT.PlotSettings()
```

```python
NReal=1000; Scale=5
NReal_R=NReal/Scale;
mean = (1, 2)
cov = [[1, 0.8], [0.8, 1]]
Input = np.random.multivariate_normal(mean, cov, 1000)
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(30,30), dpi=750, facecolor='w', edgecolor='k')

outergrid = gridspec.GridSpec(2,2,wspace=0.05,hspace=0.05,width_ratios=[1,1],height_ratios=[1,1])

color=['gray','green','red']
fontsize=35;
fontsize_T=45;
alpha=0.4;
S=200
wspace=0.02;hspace=0.02
nbins=10

# Plot 1
#==========================================
g_s = gridspec.GridSpecFromSubplotSpec(2, 2, width_ratios=[4, 1], height_ratios = [1, 4],
                                      subplot_spec=outergrid[0],wspace=wspace,hspace=hspace)
ax = plt.subplot(g_s[1, 0])
ax.scatter(Input[:,0], Input[:,1], s=S, color=color[0])
PT.BV_Plot(Input[:,0], Input[:,1], BW=-1, cmap='hot_r', Contour=True, ax=ax,alpha=alpha,fontsize=fontsize,
          cbar=False, stat_xy=(0.2, 0.99))
ax.spines['right'].set_visible(False); ax.spines['top'].set_visible(False)
ax.set_xlabel('Variable 1',fontsize=fontsize); ax.set_ylabel('Variable 2',fontsize=fontsize);

axr = plt.subplot(g_s[1, 1], sharey=ax, frameon = False, xlim=(0, 2))
axr.set_xticklabels([]);axr.set_yticklabels([])
axr.hist(Input[:,0], color = color[0], orientation = 'horizontal', normed = True, bins=nbins)

axt = plt.subplot(g_s[0, 0], sharex=ax, frameon = False, ylim=(0, 2))
axt.set_xticklabels([]);axt.set_yticklabels([])
axt.hist(Input[:,1], color = color[0], normed = True, bins=nbins)
axt.set_title('Reference',fontsize=fontsize_T,y=0.3,color=color[0])
ax.set_xlim([min(Input[:,0]), max(Input[:,0])])
ax.set_ylim([min(Input[:,1]), max(Input[:,1])])
```
